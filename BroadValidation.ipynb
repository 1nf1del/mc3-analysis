```python
germline_data = []
validated_data = []
unvalidated_data = []

val_order = [
    "mutval_status_targeted",
    "mutval_status_wgs",
    #"mutval_status_rna",
    "mutval_status_lowpass"
]

def row_extract(row):
    out = {}
    for i in ["FILTER", 'Variant_Type', 'CENTERS']:
        out[i] = row[i]
    return out

data_types = ['targeted','wgs']

def validation_status(line):
    consensus_judgement = 0
    powered = False
    for d in data_types:
        power_field = 'validation_power_' + d
        d_power = line[power_field]
        if d_power != "" and float(d_power) < 0.95:
            continue
        judgement_field = 'validation_judgement_' + d
        if line[judgement_field] == '' or line[judgement_field] == 'NaN':
            continue
        powered = True

        d_judgement = int(line[judgement_field])
        # 0 = unvalidated, 1 = validated, 2 = germline
        if d_judgement > consensus_judgement:
            consensus_judgement = d_judgement
    if not powered:
        return None
    if consensus_judgement == 0:
        return "unvalidated_powered"
    if consensus_judgement == 1:
        return "validated_powered"
    if consensus_judgement == 2:
        return "germline_powered"

with open(validation_file) as handle:
    reader = csv.DictReader(handle, delimiter="\t")
    for row in reader:
        if row['FILTER'].count("bitgt") == 0:
            s = validation_status(row)
            if s == "germline_powered":
                germline_data.append(row_extract(row))
            if s == "validated_powered":
                validated_data.append(row_extract(row))
            if s == "unvalidated_powered":
                unvalidated_data.append(row_extract(row))
```

```python
snp_callers = ["SOMATICSNIPER", "RADIA", "MUTECT", "MUSE", "VARSCANS"]
indel_callers = ["PINDEL", "INDELOCATOR", "VARSCANS"]
```

```python
def maf_row_valid(row, exons_only=False, snp=True):
    if snp:
        if row['Variant_Type'] != 'SNP':
            return False
    else:
        if row['Variant_Type'] == 'SNP':
            return False
    f = row['FILTER'].split(",")
    if exons_only:
        if "NonExonic" in f or "bitgt" in f:
            return False
    for i in ['gapfiller', 'wga', 'contest', 'native_wga_mix', 'NonExonic', 'nonpreferredpair']:
        if i in f:
            return False
    return True

def maf_row_cutoff(row, n=2):
    return len(row["CENTERS"].split("|")) >= n
    """
    count = 0
    for c in row["CENTERS"].split("|"):
        c = c.replace("*", "")
        if c in ["SOMATICSNIPER", "RADIA", "MUTECT", "MUSE", "VARSCANS"]:
            count += 1
    return count >= n
    """

def maf_row_caller(row, caller):
    s = row["CENTERS"].split("|")
    return caller in s

def maf_row_filter(row, filter):
    if filter == "NONE":
        return True
    s = row['FILTER'].split(',')
    if filter == "PASS":
        if row['FILTER'] == "PASS":
            return True
        if "NonExonic" in s:
            f.remove("NonExonic")
        if "bitgt" in s:
            f.remove("bitgt")
        if len(s) == 0:
            return True
        return False
    return filter not in s
```

```python
def get_stats(germline_data, unvalidated_data, validated_data, valid, compare):
    cfm = {
        "fp" : 0.0,
        "fn" : 0.0,
        "tp" : 0.0,
        "tn" : 0.0
    }
    germline_count = 0
    validated_count = 0
    unvalidated_count = 0
    missed_count = 0
    for row in germline_data:
        if valid(row):
            if compare(row):
                cfm["fp"] += 1
                germline_count += 1
            else:
                cfm["tn"] += 1
    for row in unvalidated_data:
        if valid(row):
            if compare(row):
                cfm["fp"] += 1
                unvalidated_count += 1
            else:
                cfm["tn"] += 1
    for row in validated_data:
        if valid(row):
            if compare(row):
                cfm["tp"] += 1
                validated_count += 1
            else:
                cfm["fn"] += 1
                missed_count += 1
    return {
        "germline_count" : germline_count,
        "validated_count" : validated_count,
        "unvalidated_count" : unvalidated_count,
        "missed_count" : missed_count,
        "matrix" : cfm
    }
   
```

```python
isect_snp_data = []
isect_indel_data = []
    
for i in range(1, 6):
    snp_data = get_stats(germline_data, unvalidated_data, validated_data, 
              lambda x: maf_row_valid(x, True, True), lambda x: maf_row_cutoff(x,i) and maf_row_filter(x,"PASS"))
    isect_snp_data.append( ("%d" % i, snp_data) )

for i in range(1, 4):
    indel_data = get_stats(germline_data, unvalidated_data, validated_data, 
          lambda x: maf_row_valid(x, True, False), lambda x: maf_row_cutoff(x,i) and maf_row_filter(x,"PASS"))
    isect_indel_data.append( ("%d" % i, indel_data) )
```

```python
caller_snp_data = []

for c in snp_callers:
    data = get_stats(germline_data, unvalidated_data, validated_data, 
              lambda x: maf_row_valid(x, True, True), lambda x: maf_row_caller(x,c) and maf_row_filter(x,"PASS"))
    caller_snp_data.append( (c, data) )
```

```python
caller_isect = []
for i, a in enumerate(snp_callers):
    for j, b in enumerate(snp_callers):
        if i > j:
            data = get_stats(germline_data, unvalidated_data, validated_data, 
              lambda x: maf_row_valid(x, True, True), lambda x: maf_row_caller(x,a) and maf_row_caller(x,b))
            caller_isect.append( ("%s and %s" % (a,b), data) )

caller_isect.sort(key=lambda x: x[1]['validated_count'], reverse=True)        
```

```python
filter_order = [
    'NONE',
    'PASS',
    'broad_PoN_v2',
    #'NonExonic',
    'StrandBias',
    #'badseq',
    #'bitgt',
    'common_in_exac',
    #'contest',
    'ndp',
    #'nonpreferredpair',
    'oxog',
    #'wga'
]

filter_data = []
for f in filter_order:
    data = get_stats(germline_data, unvalidated_data, validated_data, 
              lambda x: maf_row_valid(x, True, True), lambda x: maf_row_filter(x,f) )
    filter_data.append( (f, data) )
filter_data.sort(key=lambda x: x[1]['validated_count'], reverse=True)        
```

```python
def make_bars(datasets):
    validated = []
    germline = []
    unvalidated = []
    false_negative = []
    true_negative = []
    
    for d in datasets:
        validated.append( d['validated_count'] )
        unvalidated.append( d['unvalidated_count'] )
        germline.append( d['germline_count'] )
        false_negative.append( d['matrix']['fn'] )
        true_negative.append( d['matrix']['tn'] )
    return {
        'validated' : np.array(validated),
        'unvalidated' : np.array(unvalidated),
        'germline' : np.array(germline),
        'false_negative' : np.array(false_negative),
        'true_negative' : np.array(true_negative)
    }
```
