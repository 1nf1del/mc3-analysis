```python
%matplotlib inline
import csv
import gzip
import matplotlib.pyplot as plt
import numpy as np
import itertools
import pandas
import json
from sklearn import linear_model, model_selection, metrics
from matplotlib.backends.backend_pdf import PdfPages
from adjustText import adjust_text
```

```python
validation_file = "mc3.v0.2.8.validated.maf.gz"
```

```python
def row_id(row):
    return "%s:%s:%s:%s" % (row['Tumor_Sample_Barcode'], 
                            row['Chromosome'], 
                            row['Start_Position'],
                            row['End_Position'])
```

```python
maf_map = {}
somatic_data = []
germline_data = []
none_data = []
callers = set()
filters = set()
with gzip.GzipFile(validation_file) as handle:
    reader = csv.DictReader(handle, delimiter="\t")
    for row in reader:
        i = row_id(row)
        if i not in maf_map:
            maf_map[i] = []
        maf_map[i].append(row)

for v in maf_map.values():
    if len(v) == 1: #no ambiguous calls
        row = v[0]
        if row["VALIDATION_STATUS"] == "Somatic":
            somatic_data.append(row)
        elif row["VALIDATION_STATUS"] == "Germline":
            germline_data.append(row)
        elif row["VALIDATION_STATUS"] == "None":
            none_data.append(row)
        else:
            raise Exception("Weird Data")
        for i in row["CENTERS"].split("|"):
            callers.add(i.replace("*", ""))
        for i in row["FILTER"].split(","):
            filters.add(i)
```

```python
def filter_center_count(rows, count):
    o = []
    for r in rows:
        if len(r["CENTERS"].split("|")) >= count:
            o.append(r)
    return o
```

```python
def maf_row_method(row, method, filter=None):
    if method == "ANY":
        if filter is None:
            return True
        for f in row["FILTER"].split(","):
            if f == filter:
                return False
        return True
    centers = row["CENTERS"].split("|")
    if method in ["2", "3", "4", "5"]:
        if len(centers) >= int(method):
            if filter is None:
                return True
            for f in row["FILTER"].split(","):
                if f == filter:
                    return False
            return True
        return False        
    for c in centers:
        c = c.replace("*", "")
        if c == method:
            if filter is None:
                return True
            for f in row["FILTER"].split(","):
                if f == filter:
                    return False
            return True
    return False
```

```python
def maf_row_cutoff(row, n=2):
    count = 0
    for c in row["CENTERS"].split("|"):
        c = c.replace("*", "")
        if c in ["SOMATICSNIPER", "RADIA", "MUTECT", "MUSE", "VARSCANS"]:
            count += 1
    return count >= n
```

```python
somatic_counts = []
germline_counts = []
none_counts = []
for i in range(1, 6):
    somatic_counts.append( len(filter_center_count(somatic_data,i)) )
    germline_counts.append( len(filter_center_count(germline_data,i)) )
    none_counts.append( len(filter_center_count(none_data,i)) )
somatic_counts = np.array(somatic_counts)
germline_count = np.array(germline_counts)
none_counts = np.array(none_counts)
```

# Learn simple combined classifer

```python
def line2vector(line):
    out = {}
    #for c in ['MUTECT', 'MUSE', 'RADIA', 'SOMATICSNIPER', 'VARSCANS', 'VARSCANI', 'PINDEL', 'INDELOCATOR']:
    for c in ['MUTECT', 'MUSE', 'RADIA', 'SOMATICSNIPER', 'VARSCANS']:
        out[c] = int(c in line['CENTERS'])
    return out

learning_data = pandas.DataFrame(
    list(dict(line2vector(a), status="germline") for a in germline_data if a['Variant_Type'] == "SNP") +
    list(dict(line2vector(a), status="somatic") for a in somatic_data if a['Variant_Type'] == "SNP") +
    list(dict(line2vector(a), status="none") for a in none_data if a['Variant_Type'] == "SNP")
)
```

```python
features = learning_data.drop("status", 1)
labels = learning_data["status"].replace(["none", "germline"], 0).replace("somatic", 1)
```

```python
method_labels = ["ANY", "2", "3", "4"]
filter_labels = [
    'broad_PoN_v2',
    #'NonExonic',
    'StrandBias',
    #'bitgt',
    'common_in_exac',
    'contest',
    'ndp',
    'oxog'
]
```

```python
method_curve = []
for met in method_labels:
    cfm = {
        "fp" : 0.0,
        "fn" : 0.0,
        "tp" : 0.0,
        "tn" : 0.0
    }
    for row in germline_data:
        if row['Variant_Type'] == "SNP":
            if maf_row_method(row, met):
                cfm["fp"] += 1
            else:
                cfm["tn"] += 1
    for row in none_data:
        if row['Variant_Type'] == "SNP":
            if maf_row_method(row, met):
                cfm["fp"] += 1
            else:
                cfm["tn"] += 1
    for row in somatic_data:
        if row['Variant_Type'] == "SNP":
            if maf_row_method(row, met):
                cfm["tp"] += 1
            else:
                cfm["fn"] += 1

    fpr = float(cfm["fp"]) / float( cfm["fp"] + cfm["fn"] )
    tpr = float(cfm["tp"]) / float( cfm["tp"] + cfm["fn"] )
    fdr = float(cfm["fp"]) / float( cfm["fp"] + cfm["tp"] )
    f1 = (2.0 * cfm["tp"]) / (2.0 * cfm["tp"] + cfm["fp"] + cfm["fn"])
    
    print met, fpr, tpr
    method_curve.append((fpr, tpr))
print method_curve
```

```python
mf_curve = []
mf_labels = []

for met in method_labels:
    for filt in filter_labels:
        cfm = {
            "fp" : 0.0,
            "fn" : 0.0,
            "tp" : 0.0,
            "tn" : 0.0
        }
        for row in germline_data:
            if row['Variant_Type'] == "SNP":
                if maf_row_method(row, met, filt):
                    cfm["fp"] += 1
                else:
                    cfm["tn"] += 1
        for row in none_data:
            if row['Variant_Type'] == "SNP":
                if maf_row_method(row, met, filt):
                    cfm["fp"] += 1
                else:
                    cfm["tn"] += 1
        for row in somatic_data:
            if row['Variant_Type'] == "SNP":
                if maf_row_method(row, met, filt):
                    cfm["tp"] += 1
                else:
                    cfm["fn"] += 1
    
        fpr = float(cfm["fp"]) / float( cfm["fp"] + cfm["fn"] )
        tpr = float(cfm["tp"]) / float( cfm["tp"] + cfm["fn"] )
        fdr = float(cfm["fp"]) / float( cfm["fp"] + cfm["tp"] )
        f1 = (2.0 * cfm["tp"]) / (2.0 * cfm["tp"] + cfm["fp"] + cfm["fn"])

        
        print met, fpr, tpr
        mf_curve.append((fpr, tpr))
        mf_labels.append( "%s/%s" % (met, filt))
print mf_curve
```

```python
cutoff_curve = []
for cut in [1,2,3,4,5]:
    cfm = {
        "fp" : 0.0,
        "fn" : 0.0,
        "tp" : 0.0,
        "tn" : 0.0
    }
    for row in germline_data:
        if row['Variant_Type'] == "SNP":
            if maf_row_cutoff(row, cut):
                cfm["fp"] += 1
            else:
                cfm["tn"] += 1
    for row in none_data:
        if row['Variant_Type'] == "SNP":
            if maf_row_cutoff(row, cut):
                cfm["fp"] += 1
            else:
                cfm["tn"] += 1
    for row in somatic_data:
        if row['Variant_Type'] == "SNP":
            if maf_row_cutoff(row, cut):
                cfm["tp"] += 1
            else:
                cfm["fn"] += 1

    fpr = float(cfm["fp"]) / float( cfm["fp"] + cfm["fn"] )
    tpr = float(cfm["tp"]) / float( cfm["tp"] + cfm["fn"] )
    fdr = float(cfm["fp"]) / float( cfm["fp"] + cfm["tp"] )
    f1 = (2.0 * cfm["tp"]) / (2.0 * cfm["tp"] + cfm["fp"] + cfm["fn"])
    
    print cut, fpr, tpr
    cutoff_curve.append((fpr, tpr))
```

```python
method_marker = {
    "ALL" : "o",
    'broad_PoN_v2' : "v",
    'common_in_exac' : "^",
    'NonExonic' : "h",
    'StrandBias' : "s",
    'bitgt' : "*",
    'contest' : "D",
    'ndp' : "+",
    'oxog' : "|"
}
```
