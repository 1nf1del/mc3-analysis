```python
projects = list( a.split("/")[2].replace("_cleaned.maf", "") for a in glob("Data/somatic_mafs_cleaned/*.maf") )
```

```python
def row_id(row):
    return "%s:%s:%s:%s" % (row['Tumor_Sample_Barcode'], 
                            row['Chromosome'], 
                            row['Start_Position'],
                            row['End_Position'])

def row_extract(row):
    out = {}
    for i in ["FILTER", 'Variant_Type', 'CENTERS', "mutval_status_targeted",
    "mutval_status_wxs", "mutval_status_wgs", "mutval_status_rna", "mutval_status_lowpass"]:
        if i in row:
            out[i] = row[i]
    return out
```

```python
awg_calls = {}
for c in projects:
    awg_calls[c] = {}
    with open("Data/somatic_mafs_cleaned/%s_cleaned.maf" % (c)) as handle:
        reader = csv.DictReader(handle, delimiter="\t")
        for row in reader:
            if row['Variant_Type'] == "SNP":
                sample = row['Tumor_Sample_Barcode']
                if sample not in awg_calls[c]:
                    awg_calls[c][sample] = {}
                awg_calls[c][sample][ row_id(row) ] = row_extract(row)
```

```python
validation_file = "Data/mc3_all_pairsets.aggregated.maf"
```

```python
validation_calls = {}
with open(validation_file) as handle:
    reader = csv.DictReader(handle, delimiter="\t")
    for row in reader:
        if row['Variant_Type'] == "SNP":
            sample = row['Tumor_Sample_Barcode']
            for proj in awg_calls:
                if sample in awg_calls[proj]:
                    if proj not in validation_calls:
                        validation_calls[proj] = {}
                    if sample not in validation_calls[proj]:
                        validation_calls[proj][sample] = {}
                    validation_calls[proj][sample][ row_id(row) ] = row_extract(row)
```

```python
public_file = "Data/mc3.v0.2.8.PUBLIC.maf.gz"
public_calls = {}
with gzip.GzipFile(public_file) as handle:
    reader = csv.DictReader(handle, delimiter="\t")
    for row in reader:
        if row['Variant_Type'] == "SNP":
            sample = row['Tumor_Sample_Barcode']
            for proj in awg_calls:
                if proj not in public_calls:
                    public_calls[proj] = {}
                if sample in awg_calls[proj]:
                    if sample not in public_calls[proj]:
                        public_calls[proj][sample] = {}
                    public_calls[proj][sample][ row_id(row) ] = row_extract(row)
```
