```python
%matplotlib inline
import csv
import gzip
import matplotlib.pyplot as plt
import numpy as np
import itertools
import pandas
import json
```

```python
validation_file = "Data/mc3_all_pairsets.aggregated.maf"
```

```python
germline_data = []
validated_data = []
unvalidated_data = []

val_order = [
    "mutval_status_targeted",
    "mutval_status_wgs",
    #"mutval_status_rna",
    "mutval_status_lowpass"
]

def row_extract(row):
    out = {}
    for i in ["FILTER", 'Variant_Type', 'CENTERS']:
        out[i] = row[i]
    return out

with open(validation_file) as handle:
    reader = csv.DictReader(handle, delimiter="\t")
    for row in reader:
        if row['FILTER'].count("bitgt") == 0:
            for v in val_order:
                s = row[v]
                if s.endswith("_powered"):
                    if s == "germline_powered":
                        germline_data.append(row_extract(row))
                    if s == "validated_powered":
                        validated_data.append(row_extract(row))
                    if s == "unvalidated_powered":
                        unvalidated_data.append(row_extract(row))
                    break
```

```python
snp_callers = ["SOMATICSNIPER", "RADIA", "MUTECT", "MUSE", "VARSCANS"]
indel_callers = ["PINDEL", "INDELOCATOR", "VARSCANS"]
```

```python
def maf_row_valid(row, exons_only=False, snp=True):
    if snp:
        if row['Variant_Type'] != 'SNP':
            return False
    else:
        if row['Variant_Type'] == 'SNP':
            return False
    f = row['FILTER'].split(",")
    if exons_only:
        if "NonExonic" in f or "bitgt" in f:
            return False
    for i in ['gapfiller', 'wga', 'contest', 'native_wga_mix', 'NonExonic', 'nonpreferredpair']:
        if i in f:
            return False
    return True

def maf_row_cutoff(row, n=2):
    return len(row["CENTERS"].split("|")) >= n
    """
    count = 0
    for c in row["CENTERS"].split("|"):
        c = c.replace("*", "")
        if c in ["SOMATICSNIPER", "RADIA", "MUTECT", "MUSE", "VARSCANS"]:
            count += 1
    return count >= n
    """

def maf_row_caller(row, caller):
    s = row["CENTERS"].split("|")
    return caller in s

def maf_row_filter(row, filter):
    if filter == "NONE":
        return True
    s = row['FILTER'].split(',')
    if filter == "PASS":
        if row['FILTER'] == "PASS":
            return True
        if "NonExonic" in s:
            f.remove("NonExonic")
        if "bitgt" in s:
            f.remove("bitgt")
        if len(s) == 0:
            return True
        return False
    return filter not in s
```

```python
def get_stats(germline_data, unvalidated_data, validated_data, valid, compare):
    cfm = {
        "fp" : 0.0,
        "fn" : 0.0,
        "tp" : 0.0,
        "tn" : 0.0
    }
    germline_count = 0
    validated_count = 0
    unvalidated_count = 0
    missed_count = 0
    for row in germline_data:
        if valid(row):
            if compare(row):
                cfm["fp"] += 1
                germline_count += 1
            else:
                cfm["tn"] += 1
    for row in unvalidated_data:
        if valid(row):
            if compare(row):
                cfm["fp"] += 1
                unvalidated_count += 1
            else:
                cfm["tn"] += 1
    for row in validated_data:
        if valid(row):
            if compare(row):
                cfm["tp"] += 1
                validated_count += 1
            else:
                cfm["fn"] += 1
                missed_count += 1
    return {
        "germline_count" : germline_count,
        "validated_count" : validated_count,
        "unvalidated_count" : unvalidated_count,
        "missed_count" : missed_count,
        "matrix" : cfm
    }
```

```python

caller_order = [
    'MUTECT',
    'MUSE',
    'RADIA',
    'SOMATICSNIPER',
    'VARSCANS',
    #'VARSCANI',
    #'PINDEL',
    #'INDELOCATOR'
]

filter_order = [
    'NONE',
    'PASS',
    'broad_PoN_v2',
    #'NonExonic',
    'StrandBias',
    #'badseq',
    #'bitgt',
    'common_in_exac',
    #'contest',
    'ndp',
    #'nonpreferredpair',
    'oxog',
    #'wga'
]

rows = []
for i, caller in enumerate(caller_order):
    for j, filter in enumerate(filter_order):
        
        snp_data = get_stats(germline_data, unvalidated_data, validated_data, 
              lambda x: maf_row_valid(x, True, True), lambda x: maf_row_cutoff(x,i) and maf_row_filter(x,filter))
        
        fdr = float(snp_data['matrix']['fp']) / float(snp_data['matrix']['fp'] + snp_data['matrix']['tp'] )
        tpr = float(snp_data['matrix']['tp']) / float( snp_data['matrix']['tp'] + snp_data['matrix']['fn'] )
        fpr = float(snp_data['matrix']['fp']) / float(snp_data['matrix']['fp'] + snp_data['matrix']['fn'])
        f1 = (2.0 * snp_data['matrix']['tp']) / (2.0 * snp_data['matrix']['tp'] + snp_data['matrix']['fp'] + snp_data['matrix']['fn'] )
        rows.append( {"Caller" : caller, "Filter" : filter,  
                     "Somatic" : snp_data['validated_count'], "Germline" : snp_data['germline_count'], 
                     "FDR" : fdr, "TruePositiveRate" : tpr,
                     "FalsePositiveRate" : fpr,
                     "F1" : f1,
                     "TruePositiveCount" : snp_data["matrix"]["tp"],
                     "FalsePositiveCount" : snp_data["matrix"]["fp"],
                     "TrueNegativeCount" : snp_data["matrix"]["tn"],
                     "FalseNegativeCount" : snp_data["matrix"]["fn"],
                    })
stats = pandas.DataFrame(rows)[["Caller", "Filter", "Somatic", "Germline", 
                        "TruePositiveRate", "FalsePositiveRate", "FDR", "F1", 
                        "TruePositiveCount", "FalsePositiveCount", "TrueNegativeCount", "FalseNegativeCount" ]]#.sort_values("f1", ascending=False)
```

```python
stats.to_csv("validation_stats_snp.tsv", sep="\t")
```
